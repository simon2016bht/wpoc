```python
service = webdriver.ChromeService(executable_path = r'C:\Users\Simon\Documents\GitHub\wpoc\code\chromedriver.exe')
# service = webdriver.ChromeService(executable_path = r'C:\Users\chia-jung.s\Documents\GitHub\wpoc\code\chromedriver.exe')
driver = webdriver.Chrome(service=service)
waiting_time = 5
```

```python
links = []
start = '2025-03-01'
end = '2025-04-01'
#set links for 10 pages, which is top 1000 games
links.append(f"https://boardgamegeek.com/plays/bygame/start/{start}/end/{end}")
for i in range(2,11):
    links.append(f"https://boardgamegeek.com/plays/bygame/start/{start}/end/{end}/page/{i}")
```

```python
for page in links:
    driver.get(page)

    table = driver.find_elements(By.CLASS_NAME, "forum_table")[1]
    rows = table.find_elements(By.TAG_NAME, "tr")
    for i in rows:
        cells = i.find_elements(By.TAG_NAME, "td")
        for j in cells:
            print(j.text, end = "|||")
        try:
            game_link = cells[0].find_element(By.TAG_NAME, "a").get_attribute("href")   
            print(game_link)
        except:
            pass

    time.sleep(waiting_time)
```
