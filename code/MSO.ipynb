```python
NumPlayer = []
Titles = []

for page in links:
    driver.get(page)
    rows = driver.find_elements(By.TAG_NAME,"tr")
    title = driver.find_elements(By.CLASS_NAME,"entry-title")[0].text
    NumPlayer.append(rows[-1].find_elements(By.TAG_NAME,"td")[0].text)
    Titles.append(title)    
    time.sleep(3)
```
